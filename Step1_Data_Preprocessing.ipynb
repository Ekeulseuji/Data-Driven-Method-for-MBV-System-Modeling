```python
sampled_data = []
```

```python
import random

for i in range(89):
    for j in range(4):
        sampled = False # ç®—è¿™å››ä¸ªé‡Œé¢æœ‰æ²¡æœ‰æ•°æ®è¢«é‡‡æ ·åˆ°ï¼Œå¦‚æžœæœ€åŽä¸ºFalseçš„è¯å°±æŠŠç¬¬å››ä¸ªæ•°é‡‡æ ·è¿›åŽ»
        
        random_num = random.randint(1,10)
        if random_num <= 4:
            sampled = True
            sampled_data.append(data_list[i*4+j])
        
        # å¦‚æžœæœ€åŽä¸ºFalseçš„è¯å°±æŠŠç¬¬å››ä¸ªæ•°é‡‡æ ·è¿›åŽ»
        if j == 3 and sampled == False:
            sampled_data.append(data_list[i*4+j])
```

```python
# æœ€åŽäº”ä¸ªå¥½åƒæ²¡é‡‡åˆ°å•Šï¼Œå•å¼€ä¸€è½®
for i in range(356, 359):
    random_num = random.randint(1,10)
    print(random_num)
    if random_num <= 4:
        sampled = True
        sampled_data.append(data_list[i])
        
    # å¦‚æžœæœ€åŽä¸ºFalseçš„è¯å°±æŠŠç¬¬äº”ä¸ªæ•°é‡‡æ ·è¿›åŽ»
    if i == 359 and sampled == False:
        print(data_list[i])
        sampled_data.append(data_list[i])
```

```python
sampled_X = []
sampled_t = []

# æ„Ÿè§‰æœ‰ç‚¹å¤šæ­¤ä¸€ä¸¾ï¼ˆï¼Ÿï¼‰
for elem in sampled_data:
    sampled_X.append(elem[1])
    sampled_t.append(elem[0])
        
len(sampled_X)
```
