## 采样！

```python
# 至少每四个数里面就要有一个被采样到！
```

```python
# 创建一个空列表，用于存储一对一对应的数据
data_list = []

# 将tX一一对应存储在列表中
for i in range(360):
    data_pair = (t[i], [x_change[i], y_change[i]])
    data_list.append(data_pair)

# 打印存储在列表中的数据
print("存储在列表中的数据:")
for pair in data_list:
    print(f"Key: {pair[0]}, Value: {pair[1]}")
```

```python
sampled_data = []
```

```python
import random

for i in range(89):
    for j in range(4):
        sampled = False # 算这四个里面有没有数据被采样到，如果最后为False的话就把第四个数采样进去
        
        random_num = random.randint(1,10)
        if random_num <= 4:
            sampled = True
            sampled_data.append(data_list[i*4+j])
        
        # 如果最后为False的话就把第四个数采样进去
        if j == 3 and sampled == False:
            sampled_data.append(data_list[i*4+j])
```

```python
# 最后五个好像没采到啊，单开一轮
for i in range(356, 359):
    random_num = random.randint(1,10)
    print(random_num)
    if random_num <= 4:
        sampled = True
        sampled_data.append(data_list[i])
        
    # 如果最后为False的话就把第五个数采样进去
    if i == 359 and sampled == False:
        print(data_list[i])
        sampled_data.append(data_list[i])
```

```python
sampled_X = []
sampled_t = []

# 感觉有点多此一举（？）
for elem in sampled_data:
    sampled_X.append(elem[1])
    sampled_t.append(elem[0])
        
len(sampled_X)
```

```python
# 将X数组转换成目标形式
X_new = np.concatenate(sampled_X).reshape(-1, 2)

print(X_new)
```

```python
# 提一下数据点的x坐标和y坐标
x_values = [point[0] for point in X_new]
y_values = [point[1] for point in X_new]

plt.plot(x_values, y_values, linestyle='-')
plt.xlabel("x - Change in Carotid Artery's Internal Diameter (mm)", fontname='Times New Roman', fontsize=12)
plt.ylabel('y - Change in Mean Artery Pressure (mmHg)', fontname='Times New Roman', fontsize=12)
plt.title('Phase Diagram After Non-uniform Sampling\n', fontname='Times New Roman', fontsize=16)
plt.show()
```
